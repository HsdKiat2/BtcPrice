```python
import matplotlib.pyplot as plt
ax = price.rolling(30).mean().plot()
plt.title('Bitcoin Prices, USD')
plt.show()
```

```python
rolling = price.rolling(30).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(rolling.index.values, rolling['Price'], linewidth=2)
ticks = ['2010-10-10','2011-10-10','2012-10-10','2013-10-10','2014-10-10','2015-10-10','2016-10-10','2017-10-10','2018-10-10','2019-10-10','2020-10-10','2021-10-10']
tick_labels = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
plt.xticks(ticks, rotation='vertical')
plt.title('Bitcoin Price History')
plt.show()
```
